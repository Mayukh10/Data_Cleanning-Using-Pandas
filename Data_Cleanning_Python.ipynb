Data Cleanning in Python

```python
import pandas as pd
```

```python
df = pd.read_excel('/content/drive/MyDrive/Customer Call List.xlsx')
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
#df['Last_Name'] = df['Last_Name'].str.lstrip('...')
#df['Last_Name'] = df['Last_Name'].str.lstrip('/')
#df['Last_Name'] = df['Last_Name'].str.rstrip('/')
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
df
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
df["Phone_Number"] = df["Phone_Number"].astype(str).str.replace('Na|nan', '', regex=True)
df["Phone_Number"]
df

```

```python
df[['Street', 'City', 'State']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df['Paying Customer']= df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer']= df['Paying Customer'].str.replace('Y', 'Yes')
df['Paying Customer']= df['Paying Customer'].str.replace('No', 'N')
df['Paying Customer']= df['Paying Customer'].str.replace('N', 'No')

df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('No', 'N')
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('N', 'No')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y', 'Yes')
df
```

```python
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == '':
    df.drop(x, inplace=True)
df.fillna('', inplace=True)
df
```
